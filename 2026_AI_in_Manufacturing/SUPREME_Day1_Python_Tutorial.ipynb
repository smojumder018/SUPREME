## 11. Loops

## 12. Loop with a condition

Now we will check each temperature reading.

If the temperature is greater than 240, we will print a warning.

````python
temperatures = [220, 225, 230, 240, 250]

for temp in temperatures:
    if temp > 240:
        print(temp, "High temperature warning")
    else:
        print(temp, "Normal")
````

## 13. Functions

A function is a reusable block of code.

We use functions when we want to repeat the same task without writing the same code again and again.

````python
def check_temperature(temperature):
    if temperature > 240:
        return "Warning"
    else:
        return "Normal"

print(check_temperature(230))
print(check_temperature(250))
````

## 14. Using a function inside a loop

This is useful because we can use one function to check many sensor readings.

````python
temperatures = [220, 230, 245, 260, 235]

for temp in temperatures:
    status = check_temperature(temp)
    print("Temperature:", temp, "| Status:", status)
````

## 15. Function with two sensor inputs

Sometimes machine condition depends on more than one sensor.

Here we will use:

- temperature
- vibration

````python
def check_machine_condition(temperature, vibration):
    if temperature > 240 or vibration > 0.40:
        return "Warning"
    else:
        return "Normal"

print(check_machine_condition(230, 0.25))
print(check_machine_condition(250, 0.30))
print(check_machine_condition(265, 0.50))
````

## 16. Classes

A class is a blueprint for creating objects.

This is slightly more advanced, so we will keep it simple.

Many machine learning tools in Python use class-like objects. For example, later you may see:

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train)
model.predict(X_test)
```

For now, we will create a simple machine class.

````python
class Machine:
    def __init__(self, name, temperature, speed):
        self.name = name
        self.temperature = temperature
        self.speed = speed

    def check_status(self):
        if self.temperature > 240:
            return "Warning"
        else:
            return "Normal"

machine_1 = Machine("Machine A", 250, 1200)
machine_2 = Machine("Machine B", 230, 1000)

print(machine_1.name, "status:", machine_1.check_status())
print(machine_2.name, "status:", machine_2.check_status())
````

## 23. Fill missing values

Missing values are common in real manufacturing datasets.

For this beginner tutorial, we will use a simple method:

- Fill missing numerical values using the column mean.

This means if a temperature value is missing, we replace it with the average temperature of the temperature column.

For example:

```text
missing temperature → average temperature
```

This is called **mean imputation**.

Mean imputation is simple and easy to understand, but it is not always the best method for real research problems. More advanced methods may preserve important relationships in the data better.

For a detailed discussion of different missing data imputation techniques in additive manufacturing, you can read:

**Mojumder, S., Iqbal, N., & Halder, P. (2025).**  
*Mitigating data incompleteness in additive manufacturing using imputation techniques.*  
**Machine Learning: Engineering, 1**(1), 015011.  
https://doi.org/10.1088/3049-4761/ae2c02

In this notebook, we will start with mean imputation because our goal is to learn the basic data cleaning workflow.

````python
numeric_columns = ["temperature_C", "vibration_mm_s", "pressure_bar", "power_W"]

for col in numeric_columns:
    df[col] = df[col].fillna(df[col].mean())

df.isna().sum()
````

## 29. Check outliers using simple limits

An outlier is a value that is very different from normal values.

For this synthetic beginner dataset, we will use simple limits:

- temperature should be less than or equal to 500 °C
- vibration should be less than or equal to 2.0 mm/s
- pressure should be less than or equal to 10 bar

````python
print("Rows with very high temperature:")
display(df[df["temperature_C"] > 500])

print("Rows with very high vibration:")
display(df[df["vibration_mm_s"] > 2.0])

print("Rows with very high pressure:")
display(df[df["pressure_bar"] > 10])

print("Rows with very high power:")
display(df[df["power_W"] > 600])
````

## 30. Remove outliers

Now we will remove rows that are outside our simple limits.

In real research, we should be careful before removing outliers because an outlier can sometimes represent an important abnormal machine condition.

````python
df = df[
    (df["temperature_C"] <= 500) &
    (df["vibration_mm_s"] <= 2.0) &
    (df["pressure_bar"] <= 10) &
    (df["power_W"] <= 600)
]

df = df.reset_index(drop=True)

print("Dataset shape after removing outliers:", df.shape)
````

# Part 3: Feature Engineering

Feature engineering means creating new useful columns from existing data.

These new columns can help us better understand machine condition or train machine learning models.

## 34. Create change features

We will create:

- `temperature_change`
- `vibration_change`

These features show how much the sensor value changed compared to the previous time point.

````python
df["temperature_change"] = df["temperature_C"].diff()
df["vibration_change"] = df["vibration_mm_s"].diff()

df["temperature_change"] = df["temperature_change"].fillna(0)
df["vibration_change"] = df["vibration_change"].fillna(0)

df[["time_min", "temperature_C", "temperature_change", "vibration_mm_s", "vibration_change"]].head()
````

## 35. Create rolling average features

A rolling average smooths the data.

Instead of looking at only one reading, it looks at the average of recent readings.

Here, we will use a window of 3 time points.

````python
df["temperature_rolling_mean_3"] = df["temperature_C"].rolling(window=3, min_periods=1).mean()
df["vibration_rolling_mean_3"] = df["vibration_mm_s"].rolling(window=3, min_periods=1).mean()

df[["time_min", "temperature_C", "temperature_rolling_mean_3", "vibration_mm_s", "vibration_rolling_mean_3"]].head()
````

## 36. Normalize sensor values

Different sensors have different units and scales.

For example:

- temperature may be around 200–300
- vibration may be around 0.2–0.8
- pressure may be around 5
- power may be around 350

To combine these values in a simple risk score, we first normalize them to a 0–1 range.

````python
def min_max_normalize(series):
    return (series - series.min()) / (series.max() - series.min())

df["temperature_norm"] = min_max_normalize(df["temperature_C"])
df["vibration_norm"] = min_max_normalize(df["vibration_mm_s"])
df["pressure_norm"] = min_max_normalize(df["pressure_bar"])
df["power_norm"] = min_max_normalize(df["power_W"])

df[["temperature_norm", "vibration_norm", "pressure_norm", "power_norm"]].head()
````

## 37. Create a simple risk score

Now we will create a simple sensor-based risk score.

This is not a real industrial formula. It is only for learning.

The risk score will increase when:

- temperature is high
- vibration is high
- pressure is high
- power is high

We will give more weight to temperature and vibration because they are more directly related to machine condition in this simple example.

````python
df["risk_score"] = (
    0.40 * df["temperature_norm"]
    + 0.30 * df["vibration_norm"]
    + 0.15 * df["pressure_norm"]
    + 0.15 * df["power_norm"]
)

df[["temperature_C", "vibration_mm_s", "pressure_bar", "power_W", "risk_score"]].head()
````

# Part 4: Sensor-to-Insight

Sensor-to-insight means converting raw sensor readings into useful information.

Raw sensor readings:

```text
temperature = 251 °C
vibration = 0.65 mm/s
pressure = 5.1 bar
```

Insight:

```text
Machine condition may be risky. Maintenance check recommended.
```

This step helps engineers make decisions.

## 38. Create an insight label from risk score

We will create a new column called:

```text
maintenance_insight
```

Simple rule:

- If risk score is greater than 0.80 → `maintenance_required`
- If risk score is greater than 0.65 → `monitor_closely`
- Otherwise → `normal_operation`

````python
maintenance_insight = []

for score in df["risk_score"]:
    if score > 0.80:
        maintenance_insight.append("maintenance_required")
    elif score > 0.65:
        maintenance_insight.append("monitor_closely")
    else:
        maintenance_insight.append("normal_operation")

df["maintenance_insight"] = maintenance_insight

df[["time_min", "risk_score", "maintenance_insight"]].head(10)
````

## 39. Count insight labels

Now we will count how many rows belong to each insight category.

````python
df["maintenance_insight"].value_counts()
````

## 40. Plot risk score over time

This plot helps us see how machine risk changes over time.

````python
plt.figure(figsize=(8, 4))
plt.plot(df["time_min"], df["risk_score"], marker="o")
plt.axhline(0.65, linestyle="--", label="monitor threshold", color="orange")
plt.axhline(0.80, linestyle="--", label="maintenance threshold", color="red")
plt.xlabel("Time (min)")
plt.ylabel("Risk Score")
plt.title("Sensor-Based Risk Score Over Time")
plt.legend()
plt.grid(True)
plt.show()
````
